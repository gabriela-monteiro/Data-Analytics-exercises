```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.load_dataset("penguins")
```

```python
a = sns.load_dataset("penguins")
sns.relplot(x = "sex", y = 'body_mass_g', data = a)
```

```python
sns.catplot(x="sex", y = "body_mass_g", hue = "species", data=a, kind="strip")
```

```python
sns.lmplot(x="culmen_length_mm", y="body_mass_g", hue = "species", data=a)
```

```python
new_set = a.drop(['species', 'island', 'sex'], axis = 1)
new_set
```

```python
corr = new_set.corr()
sns.heatmap(corr)
```

```python

sns.countplot(x = 'island', hue = 'sex', data=a)
```

```python
sns.countplot(y='island', hue = 'sex', data=a)
```
